Introduction: The overall topic is safety and my submetric is the number of fire incidents. Fire is obviously an unsafe thing, so the neighborhoods with a higher amount of fire incidents will be deemed unsafe. The best neighborhood will be the one that is safest, meaning it has the least amount of fire incidents. This uses the dataset "Fire Incidents in City of Pittsburgh."

```python
import pandas as pd
import numpy as np

fire_data = pd.read_csv("https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02", index_col="_id")
```

```python
df_fire_frame = fire_data['neighborhood'].value_counts().reset_index()
df_fire_frame.columns = ['hood name', '# of incidents']
df_fire_frame
```

```python
df_fire_frame.sort_values(by = '# of incidents', ascending = False).plot.bar(x = 'hood name', figsize=(45,10))
```

Conclusion: So Carrick has the most fires, meaning you have the highest probability of experiencing a fire there, making it the most unsafe by this metric. In last place is Mount Oliver Borough, but we note there is another occurence of Mount Oliver with a higher number of incidents. This means we can ignore the lowest one, since in reality Mount Oliver has a higher amount of incidents. The best neighborhood by this metric is Regent Square.
