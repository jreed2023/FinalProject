Introduction: The overall topic is safety and my submetric is the number of fire incidents. Fire is obviously an unsafe thing, so the neighborhoods with a higher amount of fire incidents will be deemed unsafe. The best neighborhood will be the one that is safest, meaning it has the least amount of fire incidents. This uses the dataset "Fire Incidents in City of Pittsburgh."

```python
import pandas as pd
import numpy as np

fire_data = pd.read_csv("https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02", index_col="_id")
fire_data['neighborhood'].value_counts()
```

```python
#fire_data['neighborhood'].value_counts().plot.barh()
fire_data['neighborhood'].value_counts().plot.box()
```

```python
fire_data['neighborhood'].value_counts().median()
```

This box plot shows us that 50% of the neighborhoods in Pittsburgh have a fire incident number less than the median of 65. Carrick is an outlier but it should still be considered because it imapacts the result of this analysis. Outliers should be dropped when it is obvious it is measured incorrectly or contradicts assumptions being made, which Carrick does not.

```python
fire_data['neighborhood'].value_counts().plot.bar(figsize=(45,10))
```
