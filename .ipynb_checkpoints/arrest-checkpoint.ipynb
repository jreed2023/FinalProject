**Introduction**: This submetric of safety is the number of arrests. If there are more arrests in an area, then there is probably more crime there. If there is more crime, then it is less safe. The neighborhood with the least amount of arrests will be considered the best by this submetric.

```python
import pandas as pd
import numpy as np

arrest_data = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f", index_col="_id")
arrest_data["INCIDENTNEIGHBORHOOD"].value_counts()

#NOTE: LOTS OF THESE HAVE NO LISTED NEIGHBORHOOD/LAT AND LONG VALUES, SO THIS IS JUST VERY ROUGH ESTIMATE
```

```python
arrest_data['INCIDENTNEIGHBORHOOD'].value_counts().plot.box()
```

```python
arrest_data['INCIDENTNEIGHBORHOOD'].value_counts().plot.bar(figsize=(45,10))
```

**Conclusion**: Mount Oliver Neighborhood has the least amount of arrests, so by this submetric it is deemed as the safest neighborhood in the city of Pittsburgh. Conversly, the Central Business District has the highest number of arrests, so it is deemed as the least safe neighborhood.
