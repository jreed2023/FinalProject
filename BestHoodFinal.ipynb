**Introduction**
This project is to determine the best neighborhood in Pittsburgh. To figure this out we used the metric of safety because people like to live in safe places. We narrowed safety own from safety, cleanliness, and aesthetic as we believe safety should be thev priority of residents and tourists. 


To determine Pittsburgh's best/safest neighborhoood, we combined 3 different WPRDC data sub-metrics:
  - Arrests:
     The number of arrests can indicate the prevalence of criminal activity within a neighborhood. High arrest rates might suggest higher crime rates, including violent crimes, theft, or drug-related offenses. Analyzing arrest data over time can help identify trends in criminal behavior and allow law enforcement to allocate resources effectively.
  
  - Fire Incidents:
      The frequency and nature of fire incidents offer insights into public safety and the effectiveness of fire prevention measures. Higher incidents could signal issues such as inadequate fire safety infrastructure, potential hazards, or unsafe living conditions. Monitoring these incidents helps improve fire safety protocols and response times within a neighborhood.
  
  - Citations:
      Citations issued by law enforcement or other agencies reflect various infractions and violations within a community. These can range from traffic violations to code enforcement violations. Tracking citation data helps identify areas with recurring issues, such as traffic congestion, code compliance problems, or other safety concerns, allowing authorities to address them proactively.
  



**HOW SAFETY LEADS TO A BETTER NEIGHBORHOOD**  
Community Well-being: 
Safety fosters a sense of security and well-being among residents. Feeling safe in their own neighborhood allows people to lead a more relaxed and peaceful life.

Quality of Life:
A safe neighborhood enhances the overall quality of life. Families can enjoy outdoor activities, children can play freely, and residents can engage in community events without fear or concerns about their safety.

Health: 
Feeling unsafe can cause stress and anxiety, leading to negative health impacts. A safe neighborhood promotes mental and physical health by reducing stress levels and allowing people to engage in healthy activities.
Property Values: Safe neighborhoods tend to have higher property values. Potential homebuyers are often willing to pay more to live in areas perceived as safe and secure.

Community Engagement: 
Safety encourages community engagement and social interaction. Residents are more likely to participate in local activities, volunteer, and build stronger social connections when they feel secure in their environment.

Economic Development:
Businesses are more likely to invest in safe neighborhoods. A safe environment attracts commerce, which can create job opportunities and contribute to economic growth within the community.

Education: 
Safety is critical for schools within a neighborhood. A safe environment in and around schools allows students to focus on learning without fear of violence or other safety concerns.
These metrics all contribute to an overarching theme of safety. Arrests are an obvious signifier of safety as many arrestable crimes threaten the safety of a person or their property. 
  
  
  
**METRIC and how we used sub-metrics:**

Arrests:
(Pittsburgh Police Arrest Data - "Most recorded arrests are for serious offeneses, however lesser offenses also contribute")
Contains arrests made by city of Pittsburgh Police & updated daily

The first sub-metric we considered was the number of arrests in each neighborhood. If there are more arrests, that means the neighborhood is less safe because there is more crime happening. This means that the safest neighborhood by this metric is the one with the fewest number of arrests. We calculated this by simply counting the number of arrests in each neighborhood. This is not the most accurate count because many crimes did not have a neighborhood or latitude and longitude logged for them. However, there was still enough data to draw a plausible conclusion. The winner was the Mount Oliver Neighborhood and the loser was the Central Business District. 



Citations:
(Non-Traffic Citation - "Citation given for low level offenses instead of arrest")
Contains citations only given by City of Pittsburgh Police and updated daily

Next, we analyzed the non-traffic citations throughout the various Pittsburgh neighborhoods. We determined a winner and loser by adding the indivudual instances of each neighborhood. More instances in a certain neighborhood means a lower safety score wheras less instances mean a higher safety score. The winner was Mount Oliver and the loser was Southside flats.



Fires:
(Fire Incidents in City of Pittsburgh - "Incidents of fire responded by the Pittsburgh Bureau of Fire")
Recorded by block level & updated daily

Another sub-metric we looked at was the number of fire incidents in each neighborhood. Having more fires makes neighborhoods less safe because fires can harm people, so the safest neighborhood by this metric is the one with the least number of incidents. Similarly to arrests, we just had to count the number of incidents in each neighborhood and then find the minimum and maximum number and which neighborhoods they belonged to. This led us to conclude Mount Oliver Borough to be the winner and Carrick to be the loser.


**The Best Neighborhood**
To determine the overall best neighborhood, we summed the total number of all safety incidents in each one. The one with the lowest amount of total incidents was determined to be the best because people have the lowest risk of their safety being compromised there. This led to us concluding Mt Oliver as the safest and South Side Flats as the least safe.






**DATA**

```python
#Here we import all of our datasets
import pandas as pd
import numpy as np
%matplotlib inline

citation_data = pd.read_csv("https://data.wprdc.org/datastore/dump/6b11e87d-1216-463d-bbd3-37460e539d86", index_col="_id")
fire_data = pd.read_csv("https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02", index_col="_id")
arrest_data = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f", index_col="_id")
```

**ARREST DATA**

```python
df_arrest_frame = arrest_data['INCIDENTNEIGHBORHOOD'].value_counts().reset_index()
df_arrest_frame.columns = ['hood name', '# of incidents']
df_arrest_frame
```

```python
df_arrest_frame.sort_values(by = '# of incidents', ascending = False).plot.bar(x = 'hood name', figsize=(45,10))
```

*Note:* There are multiple occurances of Mount Oliver here under various different names. We will address this at the end when we sum the number of all safety incidents for each neighborhood.

**CITATION DATA**

```python
df_citation_frame = citation_data['NEIGHBORHOOD'].value_counts().reset_index()
df_citation_frame.columns = ['hood name', '# of incidents']
df_citation_frame
```

```python
df_citation_frame.sort_values(by = '# of incidents', ascending = False).plot.bar(x = 'hood name', figsize=(45,10))
```

*Note:* There are multiple occurances of Mount Oliver here under various different names. We will address this at the end when we sum the number of all safety incidents for each neighborhood.

**FIRE DATA**

```python
df_fire_frame = fire_data['neighborhood'].value_counts().reset_index()
df_fire_frame.columns = ['hood name', '# of incidents']
df_fire_frame
```

```python
df_fire_frame.sort_values(by = '# of incidents', ascending = False).plot.bar(x = 'hood name', figsize=(45,10))
```

*Note:* There are multiple occurances of Mount Oliver here under various different names. We will address this at the end when we sum the number of all safety incidents for each neighborhood.

**EVERYTHING COMBINED**

```python
df_new = pd.concat([df_arrest_frame, df_fire_frame, df_citation_frame]).groupby('hood name')['# of incidents'].sum().reset_index()
df_new.sort_values(by = '# of incidents', ascending = False)
```
