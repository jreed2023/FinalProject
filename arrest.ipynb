**Introduction**: This submetric of safety is the number of arrests. If there are more arrests in an area, then there is probably more crime there. If there is more crime, then it is less safe. The neighborhood with the least amount of arrests will be considered the best by this submetric.

```python
import pandas as pd
import numpy as np

arrest_data = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f", index_col="_id")
#NOTE: LOTS OF THESE HAVE NO LISTED NEIGHBORHOOD/LAT AND LONG VALUES, SO THIS IS JUST VERY ROUGH ESTIMATE
```

```python
df_arrest_frame = arrest_data['INCIDENTNEIGHBORHOOD'].value_counts().reset_index()
df_arrest_frame.columns = ['hood name', '# of incidents']
df_arrest_frame
```

```python
df_arrest_frame.sort_values(by = '# of incidents', ascending = False).plot.bar(x = 'hood name', figsize=(45,10))
```
